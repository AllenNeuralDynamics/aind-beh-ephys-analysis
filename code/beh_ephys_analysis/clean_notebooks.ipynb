```python
import json
from pathlib import Path

src = Path("/root/capsule/code/beh_ephys_analysis/session_combine/pupil_combine.ipynb")
dst = Path("//root/capsule/code/beh_ephys_analysis/session_combine/pupil_combine_fixed.ipynb")

nb = json.loads(src.read_text(encoding="utf-8", errors="replace"))

for c in nb.get("cells", []):
    if isinstance(c, dict) and "outputs" in c:
        c["outputs"] = []
    if isinstance(c, dict) and "execution_count" in c:
        c["execution_count"] = None

# keep backup next to original
src.with_suffix(".ipynb.bak").write_text(
    src.read_text(encoding="utf-8", errors="replace"),
    encoding="utf-8",
)

# write fixed notebook to new location
dst.parent.mkdir(parents=True, exist_ok=True)
dst.write_text(json.dumps(nb, ensure_ascii=False, indent=1), encoding="utf-8")

print(f"Done. Fixed notebook written to:\n  {dst}")
```
