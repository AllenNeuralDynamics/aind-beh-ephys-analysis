```python
from pathlib import Path
import fitz  # PyMuPDF

pdf_path = Path("/root/capsule/scratch/808655/behavior_808655_2025-09-18_13-56-51/ephys/raw/behavior_808655_2025-09-18_13-56-51_unit_beh_response.pdf")
out_dir = pdf_path.parent / f"{pdf_path.stem}_png"
out_dir.mkdir(exist_ok=True)

doc = fitz.open(pdf_path)

dpi = 300  # increase to 400 or 600 if needed
zoom = dpi / 72
mat = fitz.Matrix(zoom, zoom)

for i, page in enumerate(doc):
    pix = page.get_pixmap(matrix=mat, alpha=False)
    out_path = out_dir / f"{pdf_path.stem}_page_{i+1:03d}.png"
    pix.save(out_path)

print(f"Saved {len(doc)} PNGs to {out_dir}")
```

```python
from pathlib import Path
import fitz  # PyMuPDF

pdf_paths = [
    # Path("/root/capsule/scratch/808655/behavior_808655_2025-09-18_13-56-51/ephys/raw/figures/response/unit_100_response.pdf"),
    # Path("/root/capsule/scratch/808650/behavior_808650_2025-09-24_12-38-00/ephys/raw/figures/response/unit_254_response.pdf"),
    # Path("/root/capsule/scratch/808650/behavior_808650_2025-09-24_12-38-00/ephys/raw/figures/response/unit_100_response.pdf"),

    # Path("/root/capsule/scratch/826159/behavior_826159_2026-01-23_13-34-11/ephys/raw/figures/response/unit_42_response.pdf"),
    # Path("/root/capsule/scratch/826159/behavior_826159_2026-01-23_13-34-11/ephys/raw/figures/response/unit_171_response.pdf"),
    # Path("/root/capsule/scratch/826159/behavior_826159_2026-01-26_15-15-12/ephys/raw/figures/response/unit_25_response.pdf"),
    # Path("/root/capsule/scratch/826159/behavior_826159_2026-01-26_15-15-12/ephys/raw/figures/response/unit_39_response.pdf"),

    # Path("/root/capsule/scratch/826164/behavior_826164_2026-01-30_11-12-30/ephys/raw/figures/response/unit_87_response.pdf"),
    # Path("/root/capsule/scratch/826164/behavior_826164_2026-01-30_11-12-30/ephys/raw/figures/response/unit_95_response.pdf"),

    # Path("/root/capsule/scratch/835444/behavior_835444_2026-02-18_13-01-55/ephys/raw/figures/response/unit_47_response.pdf"),
    # Path("/root/capsule/scratch/835444/behavior_835444_2026-02-18_13-01-55/ephys/raw/figures/response/unit_202_response.pdf"),

    # Path("/root/capsule/scratch/835451/behavior_835451_2026-02-27_14-19-03/ephys/raw/figures/response/unit_115_response.pdf"),
    Path("/root/capsule/scratch/826164/behavior_826164_2026-01-28_11-12-55/ephys/raw/figures/response/unit_15_response.pdf"),
    Path("/root/capsule/scratch/826164/behavior_826164_2026-01-28_11-12-55/ephys/opto/raw/figures/unit_15_pulse_width_4_opto_tagging.pdf")
]
    

out_dir = Path("/root/capsule/scratch/PDF_to_PNG")
out_dir.mkdir(parents=True, exist_ok=True)

dpi = 300  # try 600 for larger files / higher quality
zoom = dpi / 72
mat = fitz.Matrix(zoom, zoom)

for pdf_path in pdf_paths:
    pdf_path = Path(pdf_path)

    if not pdf_path.exists():
        print(f"Missing: {pdf_path}")
        continue

    try:
        doc = fitz.open(pdf_path)

        for i, page in enumerate(doc):
            out_path = out_dir / f"{pdf_path.parent.name}_{pdf_path.stem}_page_{i+1:03d}.png"
            pix = page.get_pixmap(matrix=mat, alpha=False)
            pix.save(out_path)

        print(f"Saved {len(doc)} pages from {pdf_path.name}")

    except Exception as e:
        print(f"Failed on {pdf_path}: {e}")

print(f"\nAll PNGs saved to: {out_dir}")
```
